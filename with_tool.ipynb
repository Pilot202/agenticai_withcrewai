```python
from crewai.tools import BaseTool

class ProjectDescriptionTool(BaseTool):
    name: str = "project_description_tool"
    description: str = "Useful for finding project descriptions"

    def _run(self, message: str) -> str:
        projects = {
            "Agriscan": (
                "AgriScan is an AI-powered plant disease detection platform that helps farmers and gardeners "
                "monitor crop health. Users can upload images of their plants, and the system analyzes them "
                "using deep learning to identify diseases. It provides detailed results and saves prediction "
                "history for easy tracking. With its simple web interface, AgriScan makes early detection and "
                "crop management faster and more accessible."
            ),
            "Coinside-bot": (
                "Co-Inside is an intelligent chatbot designed to simplify learning and onboarding in the Web3 space. "
                "It provides users with personalized guidance, answering questions about blockchain, crypto, and "
                "decentralized technologies. The chatbot recommends articles, courses, and resources tailored to "
                "individual goals. By making Web3 knowledge more accessible, Co-Inside helps users confidently "
                "navigate and engage with the decentralized ecosystem."
            ),
            "moderator": (
                "The Moderator Agent is an AI system built to monitor and manage online conversations across "
                "platforms like Telegram, Discord, Slack, WhatsApp, and Facebook. It uses natural language processing "
                "to detect harmful, offensive, or unwanted content in real time. The agent can flag, tag, delete, or "
                "allow messages based on rules and AI-driven analysis. It also sends customized warnings or notifications "
                "to group members, helping communities stay safe, respectful, and well-managed."
            )
        }

        for project, description in projects.items():
            if project.lower() in message.lower():
                return description
        return "Describe this project please:"


class PersonalInformationTool(BaseTool):
    name: str = "personal_information_tool"
    description: str = (
    "Fetches the user's real personal details such as name, school, skills, GitHub username, and email. "
    "Always call this tool when you need the user's personal information for resumes, cover letters, or introductions."
)
    verbose: bool = False



    def _run(self, message: str = "") -> str:
        myname = "Aliu Usama"
        school_name = "Ladoke Akintola University of Technology"
        skillset = ["Machine Learning Engineer", "Computer Vision", "NLP", "MLOps", "Docker", "Microsoft Azure"]
        github_username = "Pilot202"
        email = "aliabiodun24@gmail.com"

        info = (
            f"My name is {myname}, I am a {school_name} student. "
            f"My skills include {', '.join(skillset)}. "
            f"My GitHub username is {github_username}, and my email is {email}."
        )

        if self.verbose:
            print(info)

        return info


# Example usage
tool1 = ProjectDescriptionTool()
print(tool1.run("Tell me about Agriscan"))

tool2 = PersonalInformationTool(verbose=True)
print(tool2.run())
```

```python
from crewai import Agent, Crew, Task
personal_info_tool = PersonalInformationTool(verbose=False)
project_tool = ProjectDescriptionTool()
email_assistant= Agent(
    role="email_assistant",
    goal="write professional emails for customers",
    backstory="I am a professional email assistant. I can write professional emails for customers.",
      tools= [personal_info_tool,project_tool],
    verbose= True,
    llm=llm

)
```

```python
original_message= " hello i would like to apply for the open machinelearning engineer role in your company, im an undergraduate student of computer engineering at the <school> my email is <EMAIL> and my github username is jonathan-m-harrison, please let me know if i can help you with anything else, thank you!, I have had hands on experience in building Agriscan, co-inside chatbot, and social media moderator app"
```

```python
email_task= Task(
    description=f" you take in the following rough message and write a professional email for a user expand abbreviation and remove dashes:'''{original_message}'''",
    agent=email_assistant,
    expected_output= "profesional email for user"
)
```

```python
email_crew= Crew(
    name="email_crew",
    agents=[email_assistant],
    tasks=[email_task],

    verbose= True
)

result= email_crew.kickoff()
print(result)
```
